```python
cur.execute("select * from owlsbrew.wfm_sales")

# TODO: RANDOMIZE STORES... maybe change store names

rownum = 0
wfm_sales = {}
for r in cur.fetchall():
    wfm_sales[rownum] = [item for item in r]
    rownum += 1

wfm_sales = DataFrame.from_dict(wfm_sales, orient = 'index')
wfm_sales.columns = 'storenum','year','month','day','sku','units','price'

wfm_sales.units = np.random.permutation(wfm_sales.units*2)

def change_sku(x):
    size = x[-1]
    prod = ''
    
    for s in x[:-1]:
        prod += s + ' '
        
    prod = prod[:-1]
    
    if prod == 'Coco-Lada':
        prod = 'Vanilla'
    elif prod == 'Pink & Black':
        prod = 'Chocolate'
    elif prod == 'The Classic':
        prod = 'Strawberry'
    else:
        prod = 'error'
    
    if size == '32OZ':
        size = '16OZ'
    elif size == '8OZ':
        size = '12OZ'
    else:
        size = 'error'
    
    return prod + ' ' + size

wfm_sales.sku = wfm_sales.loc[:,'sku'].apply(lambda x: change_sku(x.split(' ')))

wfm_sales.head()
```

```python
def insert_statement(db_name, string_list, df):
    
    insert_query = 'insert into ' + db_name + ' values'

    for i in range(len(df)):
        curr = '('
        for c in df.columns:
            if (((type(df.loc[i,c]) is numpy.float64 or\
                    type(df.loc[i,c]) is float))\
                    and math.isnan(df.loc[i,c]) or\
                    df.loc[i,c] is None):
                curr += 'NULL, '
            elif type(df.loc[i, c]) is pandas.tslib.Timestamp:
                curr += "'" + str(df.loc[i,c]) + "', "
            elif c in string_list:
                curr += "'" + str(df.loc[i,c]).replace("'","''") + "', "
            else:
                curr += str(df.loc[i,c]) + ", "
        curr = curr[:-2]
        insert_query += curr + '), '
    insert_query = insert_query[:-2]

    print insert_query
    cur.execute(insert_query)
    conn.commit()
```

## demos_new table

```python
cur.execute("select * from owlsbrew.demos_new")
rownum = 0
demos = {}
for r in cur.fetchall():
    demos[rownum] = [item for item in r]
    rownum += 1
demos = DataFrame.from_dict(demos, orient = 'index')
demos.columns = 'id','info','wfm','wfm_year','wfm_month','wfm_day','act_year','act_month','act_day','day_of_week','wfm_store'

for i in range(len(demos)):
    if demos.loc[i, 'wfm_store'] is None:
        demos = demos.drop(i)

demos.wfm_store = np.random.permutation(demos.wfm_store)

demos['delta'] = ''
for i in demos.index:
    wfm_date = date(int(demos.loc[i,'wfm_year']), int(demos.loc[i,'wfm_month']), int(demos.loc[i,'wfm_day']))
    act_date = date(int(demos.loc[i,'act_year']), int(demos.loc[i,'act_month']), int(demos.loc[i,'act_day']))
    
    demos.loc[i,'delta'] = wfm_date - act_date
    
demos.wfm_month = np.random.permutation(demos.wfm_month)    
demos.wfm_day = np.random.permutation(demos.wfm_day)

for i in demos.index:
    try:
        wfm_date = date(int(demos.loc[i,'wfm_year']), int(demos.loc[i,'wfm_month']), int(demos.loc[i,'wfm_day']))
    except: 
        wfm_date = date(int(demos.loc[i,'wfm_year']), int(demos.loc[i,'wfm_month']), 28)

    act_date = wfm_date - demos.loc[i,'delta']
    
    demos.loc[i, 'act_year'] = act_date.year
    demos.loc[i, 'act_month'] = act_date.month
    demos.loc[i, 'act_day'] = act_date.day

demos['idx'] = range(len(demos))
demos.set_index('idx', inplace = True)
demos = demos.drop('delta', 1)

create_table = 'create table case_study.demos_new(\
id int,\
info text,\
wfm tinyint,\
wfm_year int,\
wfm_month int,\
wfm_day int,\
act_year int,\
act_month int,\
act_day int,\
day_of_week text,\
wfm_store text)'

cur.execute('drop table case_study.demos_new')
cur.execute(create_table)

insert_statement('case_study.demos_new', ['info', 'day_of_week'], demos)
```
