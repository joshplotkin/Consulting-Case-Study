```python
dist = read_csv('raw_data_associated_buyers.csv')
```

```python
del dist['Unnamed: 18']
del dist['Unnamed: 19']
```

```python
def prod(x):
    x = x.split(' ')[:-1]
    name = x[0]
    for n in x[1:]:
        name += ' ' + n
    return name

dist['Product'] = dist.loc[:,'SKU'].apply(lambda x: prod(x))
dist['Size'] = dist.loc[:,'SKU'].apply(lambda x: x.split(' ')[-1])
dist
```

```python
dist.groupby('SKU').count()['SKU']
```

```python
customers = dist[['Cust #','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']].groupby('Cust #').sum().copy()
# customers.sort(ascending = False)
# customers.iloc[:6]

customers
```

```python
customers = dist.groupby('State').count()['SKU'].copy()
customers.sort(ascending = False)
customers
```

```python
months = dist[['Product','Size','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']]
months = months.groupby(['Product','Size']).sum()

months['TOTAL'] = months.sum(axis = 1)

sums = {}
sums['TOTAL'] = []
for c in months.columns:
    sums['TOTAL'].append(months.loc[:,c].sum(axis = 0))

sums = DataFrame.from_dict(sums, orient = 'index')  
sums.columns = months.columns.copy()

months = months.append(sums)
months
```
