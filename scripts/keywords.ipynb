```python
import pandas as pd
from pandas import *
```

```python
df = read_csv('./keyword/instagram_comment_keywords.csv', header = False)
```

```python
insert = 'insert into keyword(word, score, source_site, context, search_term) values '
for i in range(len(df)):
    insert += '('
    for c in df.columns:
        if c == 'score':
            insert += str(df.loc[i,c]) + ', '
        elif type(df.loc[i,c]) is float or type(df.loc[i,c]) is numpy.int64 or isnull(df.loc[i,c]):
            insert += 'NULL' + ', '
        else:
            insert += " '" + df.loc[i,c] + "', "
    insert = insert[0:-2]
    insert += '), '
insert = insert[0:-2]
print insert
```

```python
df = read_csv('./ALL_KEYWORDS2.csv')
df.head()
```

```python
df['word'] = ''
df['score'] = ''
for i in range(len(df)):
    temp = df.loc[i, 'wordscore'].split(' ')
    df.loc[i, 'score'] = temp[-1]
    tempword = temp[0:-1]
    words = []
    for w in tempword:
        if 'http' not in w:
            words.append(w.replace('\\',''))
    df.loc[i,'word'] = ' '.join(words).strip()
df = df.drop_duplicates(['site','type','key_term','search_term','word'])
df = df[['word','score','site','type','key_term','search_term']]
df
```

```python
insert = 'insert into keyword values '
for i in df.index:
    insert += '('
    for c in df.columns:
        if c == 'score':
            insert += str(df.loc[i,c]) + ', '
        elif type(df.loc[i,c]) is float or type(df.loc[i,c]) is numpy.int64 or isnull(df.loc[i,c]):
            insert += 'NULL' + ', '
        else:
            insert += " '" + df.loc[i,c] + "', "
    insert = insert[0:-2]
    insert += '), '
insert = insert[0:-2]
print insert
```
